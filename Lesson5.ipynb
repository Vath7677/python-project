```python
 # so we use this like we make a variable to store the file name and call it to use in the code 
FILE_NAME = "expenses.txt" 

# make this as lamda and back to use in (sub total)
calc_sub = lambda c, q: float(c) * int(q)

# using this as function
def data_1():
            total = 0
            try:
                # using with open because it is can close my file auto
                with open(FILE_NAME, "r") as f:
                    print("\n--- Current Records ---")
                    for line in f:
                        # .strip() we use to delete ,/speace and then 
                        # .split() it make a code share to variable and make as a list[""]
                        item, cost, qty = line.strip().split(",")

                        #after we input alredy as a string so we convert as float to calculate
                        cost_val = float(cost)
                        qty_val = int(qty)
                                    # call back lamda to use 
                        subtotal = calc_sub(cost, qty)
                        
                        print(f"- {item}: ${cost_val} x {qty_val} = ${subtotal:.2f}")
                        total += subtotal 
    
                    print("---------------------")
                    print(f"Grand Total: ${total:.2f}")
            except FileNotFoundError:
                print("Error: Data file not found!")
            except ValueError:
                print("Error: File format is incorrect. Try clearing the file (Choice 3).")
def data_2():
        num_products = int(input("How many products do you want to add?: "))
        
    # using for loop how much time i want to input
        for i in range(num_products):
            item = input(" Product Name: ")
            cost = input(" Price($): ")
            qty = input(" Quantity: ")
            
            with open(FILE_NAME, "a") as f:
                f.write(f"{item},{cost}, {qty}\n")
            print("Successfully!")
def data_3():
        confirm = input("Are u sure want to delete(yes/no): ")
        # if the input is 'yes' the data will cleaning all 
        if confirm == 'yes':
            # this mean we are not delete but we use .write to Rewrite and we can make all the thing are clear by using 
            # .write("") it mean that in the () it is have nothing so that why we seem like delete
            with open(FILE_NAME, "w") as f:
                f.write("")
            print("Data is cleaning...")
        elif confirm == 'no':
            print("Deletion cancelled.")
        else:
            print("Invalid choice...! Please type 'yes' or 'no'...")
        
    


while True:
    print("------CHOICE_MENU------")
    print("1. Checking all the data(Read)")
    print("2. Input new data(Append)")
    print("3. ReWrite all data(Write)")
    print("4. Exit")

    choice = input("Enter your choice: ")
# i call back the block of code that i use as a lamda to use and processing 
    if choice == '1':
        data_1()
    elif choice  == '2':
        data_2()
    elif choice == '3':
        data_3()
    elif choice == '4':
        print("Bye bye!")
        break
    else:
        print("Invalid choice...!")
```
