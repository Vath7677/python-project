### Initialize

```python
# constructor: to create an object class
# initialize: we use to create and prepare an object as a block of code by give them a value to use

class Dog: 
    # this is call constructor and this place mean that we tell the computer to create a plae to store an object
    def __new__(cls,name,age):            # key word '__new__' and (cls)
        return super().__new__(cls)

    
    # this is call an initialize that we use to put the value to use 
    def __init__(self, name, age):    # init key word must be '__init__' and (self) 
        self.name = name   # so we use { self. } mean that tell to create an object 
        self.age = age

dog1 = Dog("puppy", 6)
dog2 = Dog("Haski", 3)

print(f"The dog name: {dog1.name} age: {dog1.age}")
print(f"The dog name: {dog2.name} age: {dog2.age}")
```

### Inheritance

```python
# Inheritance this concept talking about how we can use other code so we can use by using { inheritance }
# so we can to make a child class can use the code in parent class

class Animal:   # parent class
    def __init__(self, name, species):
        self.name = name
        self.species = species

class Dog(Animal):  # child class
    def __init__(self, name, species, breed):  # and we can also add more item or concept to use more
        super().__init__(name, species)        # this is call form the parent class by using { super().__init__ }
        self.breed = breed                     # this is we make a new item to use in the child class


# and this we put the name and it will create an real object for use to use
my_dog = Dog("Wolf", "Mammal", "BlacK")

# this is just print an address in memory that we tell computer to create
print(my_dog)       

print(f"\n\nMy dog name: {my_dog.name}, \nspecies: {my_dog.species}, \nColor: {my_dog.breed}")
```

### Polymorphism

```python
# we use this for using the same of method name but different of action in that method

class Dog:
    def speak(self):
        return "Woof!"

class Cat:
    def speak(self):
        return "Meow!"
Animals = [Dog(),Cat()]

print(f"This is the sound of cat: {Animals[1].speak()}") # we use this for print only one 

for Animal in Animals:
    print(Animal.speak())
```

### Encapsulation 

```python
# -> we use this to make a code that very important to our program can or cannot anyone access without permission

# we for " private " we use like this: self.__balance  so it mean that { __ }  [ can use only in parentClass ]
# and for " public " we use as simple like: self.balance { can use in everywhere }
# for " protect " use like this: "self._balance" { _ }  [ u can use this in parentclass and childclass ]

class BankAccount: 
    
    # this is the same as the first one that we call initialize and just add .__ or we call private 
    def __init__(self, ID, balance):
        self.ID = ID
        self.__balance = balance

    # so we use this to see what the thing are in the balance the keyword "get_"
    def get_balance(self):
        return self.__balance

account = BankAccount("001", 100)

print(f"AccoutID: {account.ID}")
print(f"Balance: {account.get_balance()}")
        
    


```

### Magic Methods or Dunder Methods

```python
# so this is a method ot build_in key word/method that have both side of underScore {__??__}
# it is allow us to know that what we need to use in python to do what action and it is have alot

# __str__ we use to print an address that we store in memory if not have this method we cannnot print 
# this use to conver from OBJ  to print that we can read 

# __len__ we use just to count how long of the object 


class Book:
    def __init__(self,Title,page):
        self.Title = Title
        self.page = page

    def __str__(self):
        return f"Book: {self.Title}"   # we use __len__ it is not just a normal one but we use to tell 
                                        #that is how long of the book and it is the part of the book 
    def __len__(self):
        return self.page

    def __add__(self, other):            # we use this to combind it together
        return self.page + other.page
        
book1 = Book("The Hunter", 23)
book2 = Book("Python Basic", 100)

total = book1 + book2

print(book1)
print(f"The long: {len(book1)}")   
print(f"The total of the page is: {total}")               

# if i just use a simple like {my_book.page} it just a normal print 
# the python do not know u are using to to tell that is the part of the book 
```

```python
class Book:
    def __init__(self, title, author):
        self.title = title
        self.author = author
        self.__isavailable = True  

    def get_status(self):
        return "Available" if self.__isavailable else "Checked Out"

    def set_availability(self, status):
        self.__isavailable = status
        
    # using build_in function to show the thing that we can read form the memory
    def __str__(self):
        return f"'{self.title}' by {self.author} [{self.get_status()}]"

class Library:
    def __init__(self):
        self.books = []  # make a list for puting the data of book 

    def add_book(self, book):
        self.books.append(book)
        print(f"Added: {book.title}")
        
    def show_book(self):
        if not self.books:
            print("The book not in libary!!")
        else:
            for book in self.books:
                print(f"The book was in libary: {book.title}")
                
    def checkout_book(self, title):
        for book in self.books:
            if book.title.lower() == title.lower():
                if book.get_status() == "Available":
                    book.set_availability(False)
                    print(f"You checked out '{book.title}'.")
                    return
                else:
                    print(f"Sorry, '{book.title}' is already checked out.")
                    return
        print(f"Not found: '{title}'")    

    def return_book(self, title):
        for book in self.books:
            if book.title.lower() == title.lower():
                book.set_availability(True)
                print(f"Thank you! You returned '{book.title}'.")
                return
        print(f"We don't have '{title}' in our record.")

print("-----TRY OUR PROGRAM-----")

my_lib = Library()

# make a book
book1 = Book("Python 101", "Ry Vorovath")
book2 = Book("The Hunter", "Vath")

# put the book in to libarary 
my_lib.add_book(book1)
my_lib.add_book(book2)

# checking all the book in libarary 
my_lib.show_book()

# borrow hunter book 
my_lib.checkout_book("The Hunter")

# checking the book in libraray again 
my_lib.show_book()

# checking to make sure that book title have some one borrow
my_lib.checkout_book("The Hunter")

book3 = Book("Cyber security", "nha")
my_lib.add_book(book3)
my_lib.show_book()
```
